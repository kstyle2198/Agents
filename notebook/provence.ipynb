```python
from transformers import AutoModel
```

```python
provence = AutoModel.from_pretrained("D:/LLMs/provence-reranker", trust_remote_code=True)
# provence = AutoModel.from_pretrained("naver/provence-reranker-debertav3-v1", trust_remote_code=True)
print(provence)
```

```python
context = "Coco and Nana live in Seoul. They like eating fruits. Coco especially likes apple and banana. however, Nana does not like banana, instead, she likes orange and also apples. They often go to the market to buy fruits. One day, they went to the market and saw a lot of fruits. Coco bought an apple and a banana, while Nana bought an orange. They sat down on a bench and ate their fruits together. Coco shared her banana with Nana, but Nana refused it because she does not like banana."
question = 'who likes apple?'

provence_output = provence.process(question, context)
print(f"Provence Output: {provence_output}")
```
